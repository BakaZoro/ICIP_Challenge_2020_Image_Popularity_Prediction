```
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
```

```
#TRAINING DATA
X=Train_SVR[['5','3','6','7','8']].values
print(X.shape)
```

```
X[0,:]
```

```
#Labels
Y=Train_SVR.values
Y=Y[:,1]
print(Y.shape)
print(Y)
```

```
#Reshape
Y=Y.reshape((Y.shape[0],1))
```

```
#Scaling
sc_X = StandardScaler()
sc_y = StandardScaler()
Train_X = sc_X.fit_transform(X)
Train_Y = sc_y.fit_transform(Y)
```

```
Train_Y=Train_Y.reshape((Train_Y.shape[0],))
```

```
Train_Y
```

```
regressor = SVR(kernel='rbf',epsilon=0.001)
regressor.fit(Train_X,Train_Y)
```

```
#The test data
Test_X=Test_SVR[['FlickrId','5','3','6','7','8']].values
Test_Y=Test_SVR[['Views']].values
```

```
#PREDICTION
y_pred = sc_y.inverse_transform (regressor.predict (sc_X.transform(Test_X[:,1:])))
```
